```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/refs/heads/main/Dados/imigrantes_canada.csv')
```

```python
df
```

```python
df.info()
```

```python
df.set_index('País', inplace=True) #Define a coluna 'País' como índice do DataFrame df, alterando-o diretamente
```

```python
anos = list(map(str, range(1980, 2014))) #Converte cada número do intervalo de 1980 a 2013 em uma string, criando uma lista desses anos.
```

```python
anos
```

```python
brasil = df.loc['Brasil', anos] #Armazena na variável brasil os valores de todas as colunas representadas pelos anos na lista anos para a linha correspondente ao índice 'Brasil' no DataFrame
```

```python
brasil
```

```python
#Cria um dicionário chamado brasil_dict que contém duas chaves: 'Ano' e 'Imigrantes'.
brasil_dict = {'Ano': brasil.index.tolist(), 'Imigrantes': brasil.values.tolist()}

#Cria um novo DataFrame chamado dados_brasil a partir do dicionário brasil_dict.
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
#Configura o tamanho do grafico
plt.figure(figsize=(8, 4))

#Plota os dados de imigrantes ao longo dos anos no Brasil
plt.plot(dados_brasil['Ano'], dados_brasil['Imigrantes'])

#Define o titulo do grafico
plt.title('Imigração do Brasil para o Canada')

#Define o nome do eixo X e o eixo Y
plt.xlabel('Ano')
plt.ylabel('Imigrantes')

#configura os rótulos do eixo x para os anos específicos listados.
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])

#Mostra na tela apenas o grafico
plt.show()
```

```python
#Cria uma figura com 1 linha e 2 colunas de subplots
fig, axs = plt.subplots(1,2, figsize=(15, 5))

#Plota os dados do primeiro subplot
axs[0].plot(dados_brasil['Ano'], dados_brasil['Imigrantes'])

#Define o titulo do primeiro subplot
axs[0].set_title('Imigração do Brasil para o Canada\n1980 a 2013')

#define os eixos X e Y do primeiro subplot
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Numero de Imigrantes')

##Configura o eixo X do gráfico para exibir marcadores principais a cada 5 unidades
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))

#Adiciona grade no primeiro subplot
axs[0].grid()


# Cria um boxplot dos dados de imigração do Brasil para o Canadá
axs[1].boxplot(dados_brasil['Imigrantes'])

# Define o título do segundo subplot
axs[1].set_title('Boxplot da imigração do Brasil para o Canada\n1980 a 2013')

# Define os eixos X e Y do segundo subplot"
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Numero de Imigrantes')

# Configura os ticks principais do eixo x para cada 5 unidades
axs[1].xaxis.set_major_locator(plt.MultipleLocator(5))

# Ativa a grade no segundo subplot
axs[1].grid()


plt.show()
```

```python
# Cria uma figura com 2 linhas e 2 colunas de subplots, com tamanho total de 10x6 polegadas
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Ajusta o espaçamento vertical e horizontal entre os subplots
fig.subplots_adjust(hspace=0.5, wspace=0.3)

# Define o título principal da figura
fig.suptitle('Comparação de imigração dos paises para o Canada de 1980 a 2013')

# Plota os dados de imigração do Brasil no primeiro subplot e define o título
axs[0, 0].plot(df.loc['Brasil', anos])
axs[0, 0].set_title('Brasil')

# Plota os dados de imigração da Venezuela no segundo subplot e define o título
axs[0, 1].plot(df.loc['Venezuela', anos])
axs[0, 1].set_title('Venezuela')

# Plota os dados de imigração da Argentina no terceiro subplot e define o título
axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

# Plota os dados de imigração do Peru no quarto subplot e define o título
axs[1, 1].plot(df.loc['Peru', anos])
axs[1, 1].set_title('Peru')

# Define os valores mínimo e máximo para o eixo y
ymin = 0
ymax = 7000

# Configura os rótulos e a grade dos eixos x e y para todos os subplots
for ax in axs.flat:
    ax.set(xlabel='Ano', ylabel='Numero de Imigrantes')
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.grid()

# Define os limites do eixo y para todos os subplots
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

# Exibe a figura com todos os subplots configurados
plt.show()
```

```python
#Cria uma figura e um eixo com tamanho 8x4 polegadas
fig, ax = plt.subplots(figsize=(8, 4))

#plota os dados de imigrantes ao longo dos anos no Brasil no eixo, e muda a grossura da linha do grafico, e muda a cor da linha
ax.plot(dados_brasil['Ano'], dados_brasil['Imigrantes'], lw=3, color='g')

#Define o titulo da figura, muda o tamanho, e ajusta o lado que ficara
ax.set_title('Imigração do Brasil para o Canada\n1980 a 2013', fontsize=18, loc='left')

#Define o nome dos eixos X e Y
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Numero de Imigrantes', fontsize=14)

#Define o tamanho dos TICKS dos eixos X e Y
ax.tick_params(axis='both', labelsize=12)

#Configura o eixo X do gráfico para exibir marcadores principais a cada 5 unidades
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

#Mostra a figura
plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"') #Busca apenas os paises que estao na america do sul
```

```python
america_sul
```

```python
# Ordena o DataFrame 'america_sul' com base na coluna 'Total' em ordem crescente
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
# Inicializa a lista de cores vazia
cores = []

# Adiciona cores à lista com base no país
for pais in america_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('green')  # Se o país for Brasil, adiciona a cor verde
    else:
        cores.append('silver')  # Para outros países, adiciona a cor prata

# Cria uma figura e um eixo com tamanho 12x5 polegadas
fig, ax = plt.subplots(figsize=(12, 5))

# Plota um gráfico de barras horizontais com as cores definidas
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)

# Define o título do gráfico
ax.set_title('America do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canada no periodo de 1980 A 2013', loc='left', fontsize=14)

# Define o rótulo do eixo x
ax.set_xlabel('Numero de Imigrantes', fontsize=14)
ax.set_ylabel('')  # Deixa o rótulo do eixo y vazio

# Ajusta o tamanho dos rótulos dos ticks para ambos os eixos
ax.tick_params(axis='both', labelsize=12)

# Adiciona os valores das barras no gráfico
for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

# Remove todas as bordas do gráfico
ax.spines[['top', 'right', 'left', 'bottom']].set_visible(False)

# Define a visibilidade do eixo x como falso
ax.get_xaxis().set_visible(False)

# Remove os ticks dos eixos x e y
ax.tick_params(axis='both', which='both', length=0)

# Exibe o gráfico
plt.show()
```

```python
import seaborn as sns
```

```python
# Define o estilo visual padrão para os gráficos criados com a biblioteca Seaborn
sns.set_theme()
```

```python
# Ordena o DataFrame 'df' em ordem decrescente com base na coluna 'Total'
# e seleciona os 10 primeiros registros
top_10 = df.sort_values('Total', ascending=False).head(10)
```

```python
top_10
```

```python
# Cria um gráfico de barras usando a biblioteca Seaborn
# Os dados utilizados são do DataFrame 'top_10'
sns.barplot(data=top_10, x=top_10.index, y='Total')
```

```python
# Cria um gráfico de barras horizontal usando a biblioteca Seaborn
sns.barplot(
    data=top_10,        # Usa o DataFrame 'top_10' como fonte de dados
    y=top_10.index,     # Define o índice do DataFrame como categorias no eixo Y
    x='Total',          # Define os valores da coluna 'Total' como eixo X
    orient='h'          # Configura a orientação do gráfico como horizontal (h)
)
```

```python
# Cria um gráfico de barras horizontal utilizando a biblioteca Seaborn
ax = sns.barplot(
    data=top_10,         # Define o DataFrame 'top_10' como fonte dos dados
    y=top_10.index,      # Utiliza o índice do DataFrame para o eixo vertical (Y)
    x='Total',           # Utiliza a coluna 'Total' como valores no eixo horizontal (X)
    orient='h'           # Configura a orientação das barras como horizontal
)

# Configura o título e os rótulos do gráfico
ax.set(
    title='Países com mais imigrantes para o Canada\n1980 a 2013',  # Define o título do gráfico
    xlabel='Numero de imigrantes',  # Define o título do eixo X
    ylabel=''                       # Remove o título do eixo Y
)

# Exibe o gráfico na tela
plt.show()
```

```python
# Cria uma figura e um eixo com tamanho 8x4 polegadas
fig, ax = plt.subplots(figsize=(8, 4))

# Cria um gráfico de barras horizontal usando o Seaborn
ax = sns.barplot(
    data=top_10,       # Define o DataFrame 'top_10' como a fonte dos dados
    y=top_10.index,    # Define as categorias no eixo vertical (Y) a partir do índice do DataFrame
    x='Total',         # Define os valores no eixo horizontal (X) a partir da coluna 'Total'
    orient='h'         # Configura a orientação do gráfico como horizontal
)

# Define o título do gráfico com localização ajustada para a esquerda e tamanho da fonte maior
ax.set_title(
    'Países com mais imigrantes para o Canada\n1980 a 2013',  # Título do gráfico
    loc='left',        # Alinha o título à esquerda
    fontsize=16        # Define o tamanho da fonte do título
)

# Define o rótulo do eixo X
ax.set_xlabel(
    'Numero de imigrantes',  # Texto para o rótulo do eixo X
    fontsize=14              # Define o tamanho da fonte do rótulo
)

# Remove o rótulo do eixo Y (as categorias do índice já são exibidas)
ax.set_ylabel('')

# Exibe o gráfico
plt.show()
```

```python
# Define uma função que gera um gráfico de barras horizontal utilizando uma paleta personalizada
def gerar_grafico_paleta(palette):
    # Cria uma figura e um eixo com tamanho de 8x4 polegadas
    fig, ax = plt.subplots(figsize=(8, 4))

    # Plota um gráfico de barras horizontal usando o Seaborn
    # 'palette' define a paleta de cores a ser utilizada no gráfico
    ax = sns.barplot(
        data=top_10,        # Define o DataFrame 'top_10' como fonte dos dados
        y=top_10.index,     # Usa o índice do DataFrame no eixo Y (categorias)
        x='Total',          # Usa a coluna 'Total' no eixo X (valores)
        orient='h',         # Define a orientação das barras como horizontal
        palette=palette     # Aplica a paleta de cores passada como argumento da função
    )

    # Configura o título do gráfico com alinhamento à esquerda e tamanho da fonte aumentado
    ax.set_title(
        'Paises com maior imigração para o canada\n1980 A 2013',  # Título do gráfico
        loc='left',        # Alinha o título à esquerda
        fontsize=18        # Define o tamanho da fonte do título
    )

    # Define o rótulo do eixo X
    ax.set_xlabel(
        'Numero de Imigrantes',  # Texto do rótulo do eixo X
        fontsize=14              # Tamanho da fonte do rótulo
    )

    # Remove o rótulo do eixo Y
    ax.set_ylabel('')

    # Exibe o gráfico
    plt.show()
```

```python
import plotly.express as px #Importa a biblioteca plotly pra criar graficos interativos
```

```python
# Cria uma figura e um eixo com tamanho 8x4 polegadas
fig, ax = plt.subplots(figsize=(8, 4))

# Cria um gráfico de barras horizontal usando a biblioteca Seaborn
ax = sns.barplot(
    data=top_10,         # DataFrame contendo os dados do top 10
    y=top_10.index,      # Usa o índice do DataFrame para o eixo vertical (categorias)
    x='Total',           # Usa os valores da coluna 'Total' para o eixo horizontal
    orient='h',          # Configura a orientação das barras como horizontal
    palette='tab10'      # Aplica a paleta de cores 'tab10', uma paleta padrão do Seaborn
)

# Adiciona o título ao gráfico
ax.set_title(
    'Paises com maior imigração para o canada\n1980 a 2013',  # Texto do título
    loc='left',         # Alinha o título à esquerda
    fontsize=18         # Define o tamanho da fonte do título
)

# Define o rótulo do eixo X
ax.set_xlabel(
    'Numero de Imigrantes',  # Texto do rótulo do eixo X
    fontsize=14              # Define o tamanho da fonte do rótulo
)

# Remove o rótulo do eixo Y (as categorias já são exibidas no eixo Y)
ax.set_ylabel('')

# Remove as bordas desnecessárias do gráfico
sns.despine()

# Exibe o gráfico
plt.show()
```
